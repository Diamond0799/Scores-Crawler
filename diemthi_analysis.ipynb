```python
KC = df[df['Ngu Van'].notnull() & df['Lich Su'].notnull() & df['Dia Ly'].notnull()]
```

```python
KC = KC.loc[:, ['SBD', 'Ngu Van', 'Lich Su', 'Dia Ly']]
KC['C'] = KC['Ngu Van'] + KC['Lich Su'] + KC['Dia Ly']
```
